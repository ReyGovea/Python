## Question 1a

code adapted from https://www.analyticsvidhya.com/blog/2019/10/gaussian-mixture-models-clustering/

```python
x1 = pd.read_csv('/content/drive/MyDrive/Fall 2021/Applied Machine Learning /Homework /Rey and Mark Colab Work/HW10/x1.txt', sep=",", header= None)
x1.head()
```

```python
gmm = GaussianMixture(n_components=2, init_params='kmeans')
gmm.fit(x1)

#predictions from gmm
labels = gmm.predict(x1)
frame = pd.DataFrame(x1)
frame['cluster'] = labels
frame.columns = ['0', '1', 'cluster']

color=['blue','green']
for k in range(2):
    x1 = frame[frame["cluster"]==k]
    plt.scatter(x1["0"],x1["1"],c=color[k])
plt.show()
```

## Question 2
code adapted from:
https://towardsdatascience.com/unsupervised-machine-learning-spectral-clustering-algorithm-implemented-from-scratch-in-python-205c87271045

```python
#Re-initialize data
x1 = pd.read_csv('/content/drive/MyDrive/clustering/x1.txt', sep=",", header= None)
x2 = pd.read_csv('/content/drive/MyDrive/clustering/x2.txt', sep=",", header= None)
pts = pd.read_csv('/content/drive/MyDrive/clustering/pts.txt', sep=",", header= None)
```

```python
sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors')
sc_clustering = sc.fit(x1)
plt.scatter(x1.iloc[:,0], x1.iloc[:,1], c=sc_clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.show()
```

```python
sc = SpectralClustering(n_clusters=2, affinity='nearest_neighbors')
sc_clustering = sc.fit(x2)
plt.scatter(x2.iloc[:,0],x2.iloc[:,1], c=sc_clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.show()
```

```python
sc = SpectralClustering(n_clusters=3, affinity='nearest_neighbors')
sc_clustering = sc.fit(pts)
plt.scatter(pts.iloc[:,0],pts.iloc[:,1], c=sc_clustering.labels_, cmap='rainbow', alpha=0.7, edgecolors='b')
plt.show()
```
